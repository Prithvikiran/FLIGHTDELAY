```python
PATH = r"weather"
EXT = "*.json"
files = [file
                 for path, subdir, files in os.walk(PATH)
                 for file in glob(os.path.join(path, EXT))]
print(files)
```

```python
len(files)
```

```python
for i in files:
    parts = i.split(os.path.sep)
    print(parts[1])
```

```python
l = []
for i in files:
    df_1 = pd.read_json(i)
    parts = i.split(os.path.sep)
    temp = df_1["data"]["weather"]
    weather_df = pd.json_normalize(temp, record_path = 'hourly', meta = 'date')
    weather_df['ACODE'] = parts[1]
    l.append(weather_df)
 
```

```python
beta=pd.DataFrame()
beta=pd.concat(l)
beta
```
